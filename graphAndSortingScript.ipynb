```python
import csv
import operator
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt2
%matplotlib inline
count = 0
filenames = ['outputFCFS.txt', 'outputSJF.txt', 'outputRR.txt']
printFiles = ['no worries bro','outputSJFSorted.txt', 'outputRRSorted.txt']
titles = ['First Come First Serve', 'Shortest Job First', 'Round Robin']

for title in filenames:
    if(count != 0):
        sample = open(title, 'r')
        reader = csv.reader(sample, delimiter=' ')
        sort = sorted(reader, key=lambda row: int(row[0]))
        output = open(printFiles[count], 'w')
        for line in sort:
            print(line, file=output)
        
    data = np.loadtxt(title,delimiter=' ', dtype=(int))
    waiting = data[:,1]
    turnaround = data[:,2]

    plt.hist(waiting, rwidth=0.95, color='purple')
    plt.xlabel("Waiting Time")
    plt.ylabel("Number of processes")
    plt.title(titles[count])
    plt.show();


    plt2.hist(turnaround, rwidth=0.95, color='r')
    plt2.xlabel("Turnaround time")
    plt2.ylabel("Number of processes")
    plt2.title(titles[count])
    plt2.show();
    count = count + 1
```
